```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("social_media_data.csv")
data.head()
```

```python
data.fillna(0, inplace=True)
```

```python
data['platform'].value_counts()
```

```python
v_platform=["Twitter", "Instagram", "Facebook"]
data= data[data['platform'].isin(v_platform)]
```

```python
plt.bar(data['platform'], data['post_type'], color='blue')
```

```python
sns.barplot(x='platform', y='likes', data = data, color='skyblue', errorbar=None)
```

```python
plt.figure(figsize=(8,6))
data['platform'].value_counts().plot(kind='bar', color='red')
```

```python
data=data.sort_values('date')

sns.lineplot(x='date',y='views', data=data.head(20), color='teal', linestyle='--')
plt.xticks(rotation=45)
plt.show()
```

```python
data=data.head(50)
plt.plot(data['date'], data['views'])
plt.xticks(rotation=45)
plt.show()
```

```python
corr= data[['shares','views','likes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
num_df= data.select_dtypes(include='number')
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.title('HEATBLAST')
plt.xlabel('BLAST')
plt.ylabel('HEAT')

plt.show()
```

```python
sns.boxplot(x='platform', y='comments', data=data, hue='likes',palette='Set2')
plt.xticks(rotation=45);
```

```python
plt.scatter(data['views'], data['likes'], c=data['likes'], cmap='viridis')
plt.colorbar()
```
